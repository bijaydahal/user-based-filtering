```python
import pandas as pd
ratings = pd.read_csv('ml1m_ratings.csv', 
                      sep='\t', 
                      usecols=['userid', 'movieid', 'user_emb_id', 'movie_emb_id', 'rating'])
```

```python
ratings.head()
```

```python
movies = pd.read_csv('ml1m_movies.csv', 
              sep='\t',
              usecols=['movieid', 'title', 'genre'])
```

```python
movies.head()
```

```python
ratings = pd.merge(movies,ratings).drop(['genre','movie_emb_id','user_emb_id'],axis=1)
```

```python
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['userid'],columns=['title'],values='rating')
```

```python
userRatings.head()
```

```python
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
```

```python
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(10)
```

```python
corrMatrix
```

```python
corrMatrix['rating']
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
romantic_lover = [("Father of the Bride Part II (1995)",5),("Sudden Death (1995)",3),("Balto (1995)",1),("Nixon (1995)",2)]
similar_movies = pd.DataFrame()
for movie,rating in romantic_lover:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)

similar_movies.head(10)
```
